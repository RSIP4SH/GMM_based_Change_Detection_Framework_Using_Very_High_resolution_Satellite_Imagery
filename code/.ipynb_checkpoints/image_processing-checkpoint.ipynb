```python
import os
imgdir = '../dataset'
imgfile1 = os.path.join(imgdir, 'k02-05m.tif')
imgfile2 = os.path.join(imgdir, 'k12-05m.tif')
```

```python
from skimage import io
img1 = io.imread(imgfile1)
type(img1)
```

```python
print img1.shape
```

```python
img2 = io.imread(imgfile2)
print img2.shape
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure(figsize=(16,35))
plt.subplot(121).imshow(img1)
plt.subplot(122).imshow(img2)
```

```python
from skimage.transform import rotate
img1r = rotate(img1, 3.6, resize=True)
fig = plt.figure(figsize=(12,12))
plt.imshow(img1r)
```

```python
import scipy.misc
scipy.misc.imsave(os.path.join(imgdir,"k02-05m-rotated.png"), img1r)
print img1r.shape
```

```python
img2r = rotate(img2, 1.8, resize=True)
fig = plt.figure(figsize=(12,12))
plt.imshow(img2r)
```

```python
scipy.misc.imsave(os.path.join(imgdir,"k12-05m-rotated.png"), img2r)
print img2r.shape
```
